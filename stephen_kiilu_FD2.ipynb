```python
import numpy as np 
import matplotlib.pyplot as plt
import pylab as pl
```

```python
def RK4(f,c,t):
    H = t[1] - t[0]
    N = len(t)
    M = len(c)
    z = np.zeros((N,M))
    z[0] = c
    for i in range(0,N-1): ## Range Kutta
        k1 = f(z[i],t[i])
        k2 = f(z[i]+H*(k1/2),t[i]+H/2)
        k3 = f(z[i]+H*(k2/2),t[i]+H/2)
        k4 = f(z[i]+H*k3,t[i]+H)
        z[i+1] = z[i]+H/6*(k1+2*k2+2*k3+k4)
    return z
```

# Number 1.


\begin{eqnarray*}
\frac{dx}{dt}&=&\begin{pmatrix}
5 & 1 \\
3& 1 
\end{pmatrix}x
\end{eqnarray*}

```python
def fun(z,t):
    x,y=z
    return np.array([5*x+y,3*x+y])
```

```python
t=np.linspace(0,2,200)
```

```python
plt.figure(figsize = (10,8))
t1=np.linspace(0,2,200)
for x_0 in np.linspace(-2,2,12):
    for y_0 in np.linspace(-2,2,12):
        z0 = [x_0, y_0]
        z=RK4(fun,z0,t1)
        plt.plot(z[:,0],z[:,1])
        
t2=np.linspace(0,-2,200)      
for x_0 in np.linspace(-2,2,12):
    for y_0 in np.linspace(-2,2,12):
        z0 = [x_0, y_0]
        z=RK4(fun,z0,t2)
        plt.plot(z[:,0],z[:,1])
        
plt.xlim(-2, 2)
plt.ylim(-2, 2);
plt.xlabel("x")
plt.ylabel("y")
plt.title("Phase potrait of an improper node")

s,w = np.mgrid[-2:2:6j, -2:2:6j]
u = 5*s +w
v = 3*s+w

pl.quiver(s, w, u, v, color = "red")
plt.show()
```
