```python
import numpy as np 
import matplotlib.pyplot as plt
import pylab as pl
```

```python
def RK4(f,c,t):
    H = t[1] - t[0]
    N = len(t)
    M = len(c)
    z = np.zeros((N,M))
    z[0] = c
    for i in range(0,N-1): ## Range Kutta
        k1 = f(z[i],t[i])
        k2 = f(z[i]+H*(k1/2),t[i]+H/2)
        k3 = f(z[i]+H*(k2/2),t[i]+H/2)
        k4 = f(z[i]+H*k3,t[i]+H)
        z[i+1] = z[i]+H/6*(k1+2*k2+2*k3+k4)
    return z
```

```python
def fun(z,t):
    x,y=z
    return np.array([-4*x+y,x])
```

```python
t=np.linspace(0,2,200)
```

```python
plt.figure(figsize = (12,8))
t1=np.linspace(0,2,200)
for x_0 in np.linspace(-2,2,12):
    for y_0 in np.linspace(-2,2,12):
        z0 = [x_0, y_0]
        z=RK4(fun,z0,t1)
        plt.plot(z[:,0],z[:,1])
        
t2=np.linspace(0,-2,200)      
for x_0 in np.linspace(-2,2,12):
    for y_0 in np.linspace(-2,2,12):
        z0 = [x_0, y_0]
        z=RK4(f,z0,t2)
        plt.plot(z[:,0],z[:,1])
        
plt.xlim(-2, 2)
plt.ylim(-2, 2);
plt.xlabel("x")
plt.ylabel("y")
plt.title("Phase potrait")


s=np.linspace(-1,1,5)
w=np.linspace(-1,1,5)
u = -4*s +w
v = s


pl.quiver(s, w, u, v, color = "blue")
plt.show()
```
