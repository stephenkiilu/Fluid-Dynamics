# Damping and Phase diagram for different values of Beta

```python
def acceleration(x,B,v):
    return -x-B*v


def function(dt,a,b,B):
    
    t=np.arange(a,b+dt,dt)
    
    x=np.zeros(len(t))
    
    v=np.zeros(len(t))
    
    x[0]=1
    
    v[0]=1
    
    for n in range(0,len(t)-1):
        A = acceleration(x[n],B,v[n] )
        
        x[n+1] = x[n]  + dt *v[n] + (1/2)* (dt**2) * acceleration(x[n] , B, v[n])
        
        v_hat = v[n] + (1/2) * dt * (A + acceleration(x[n+1] ,B, v[n] +dt *A) )
        
        v[n+1] = v[n] + (1/2) *dt * (A+acceleration(x[n+1], B ,v_hat))
    return t,x,v
    
   
    
    
```

```python
dt =0.01
a= 0
b= 15

```

# Plot of x(t) against time

```python
Beta = np.linspace(0.01,4,6)
for i in Beta:
    t,x,v = function(dt,a,b,i)
    plt.plot(t,x,label=round(i,1))
    plt.legend()
    plt.xlabel("time")
    plt.ylabel("x(t)")
    plt.title("plot of x(t) against time")
plt.grid()
plt.show()

    
```

# The phase diagram

```python
Beta = np.linspace(0.01,4,6)
for i in Beta:
    t,x,v = function(dt,a,b,i)
    plt.plot(x,v,label=round(i,1))
    plt.legend()
    plt.xlabel("x(t)")
    plt.ylabel("v(t)")
    plt.title("plot of x(t) against v(t)")
plt.grid()
plt.show()
```

# Damping and phase diagram when you a more realistic friction force

```python
def acceleration(x,B,v):
    return -x+B*np.sin(v)


def function(dt,a,b,B):
    
    t=np.arange(a,b+dt,dt)
    
    x=np.zeros(len(t))
    
    v=np.zeros(len(t))
    
    x[0]=1
    
    v[0]=0
    
    for n in range(0,len(t)-1):
        A = acceleration(x[n],B,v[n] )
        
        x[n+1] = x[n]  + dt *v[n] + (1/2)* (dt**2) * acceleration(x[n] , B, v[n])
        
        v_hat = v[n] + (1/2) * dt * (A + acceleration(x[n+1] ,B, v[n] +dt *A) )
        
        v[n+1] = v[n] + (1/2) *dt * (A+acceleration(x[n+1], B ,v_hat))
    return t,x,v
    
```

# Plot x(t) against time

```python
Beta = np.linspace(0.01,4,6)
for i in Beta:
    t,x,v = function(dt,a,b,i)
    plt.plot(t,x,label=round(i,1))
    plt.legend()
    plt.xlabel("time")
    plt.ylabel("x(t)")
    plt.title("plot of x(t) against time")
plt.grid()
plt.show()
```

```python
Beta = np.linspace(0.01,4,6)
for i in Beta:
    t,x,v = function(dt,a,b,i)
    plt.plot(x,v,label=round(i,1))
    plt.legend()
    plt.xlabel("x(t)")
    plt.ylabel("v(t)")
    plt.title("plot of v(t) against x(t)")
plt.grid()
plt.show()
```
